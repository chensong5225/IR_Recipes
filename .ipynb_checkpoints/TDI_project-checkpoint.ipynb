```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('epi/epi_r.csv')
data.head(1)
```

```python
# remove na
df = data.dropna()
```

```python
# remove extreme values
plt.boxplot(df.calories)
```

```python
df2 = df[df.calories < 5000]
plt.boxplot(df2.calories)
```

```python
plt.boxplot(df2.sodium)
```

```python
df3 = df2[df2.sodium < 10000]
plt.boxplot(df3.sodium)
```

```python
def rm_outliers(data, m=3):
    return data[abs(data - np.mean(data)) < m * np.std(data)]

df3['calories_clean'] = rm_outliers(df3['calories'])
plt.boxplot(df3.calories_clean.dropna())
```

```python
df3['fat_clean'] = rm_outliers(df3['fat'])
plt.boxplot(df3.fat_clean.dropna())
```

```python
df3['protein_clean'] = rm_outliers(df3['protein'])
plt.boxplot(df3.protein_clean.dropna())
```

```python
df3['sodium_clean'] = rm_outliers(df3['sodium'])
plt.boxplot(df3.sodium_clean.dropna())
```

```python
df4 = df3.dropna()
len(df4)
```

```python
df5 = pd.DataFrame({'title':df4.title, 'rating':df4.rating, 'calories':df4.calories_clean, 'fat':df4.fat_clean, 
                    'sodium':df4.sodium_clean, 'protein':df4.protein_clean, 'healthy':df4.healthy})
df5.head(1)
```

```python
df_h = df5[df5.healthy == 1]
df_n = df5[df5.healthy == 0]
```

```python
plt.hist(df_h.calories, bins=10, normed=True, color='green', alpha=.9)
plt.hist(df_n.calories, bins=10, normed=True, color='red', alpha=.5)
from matplotlib.patches import Rectangle
handles = [Rectangle((0,0),1,1,color=c,ec="k") for c in ['green','red']]
labels= ['healthy','not_healthy']
plt.legend(handles, labels)
plt.title("calories")
```

```python
plt.hist(df_h.fat, bins=10, normed=True, color='green', alpha=.9)
plt.hist(df_n.fat, bins=10, normed=True, color='red', alpha=.5)
plt.legend(handles, labels)
plt.title("fat")
```

```python
plt.hist(df_h.sodium, bins=10, normed=True, color='green', alpha=.9)
plt.hist(df_n.sodium, bins=10, normed=True, color='red', alpha=.5)
plt.legend(handles, labels)
plt.title("sodium")
```

```python
plt.hist(df_h.protein, bins=10, normed=True, color='green', alpha=.9)
plt.hist(df_n.protein, bins=10, normed=True, color='red', alpha=.5)
plt.legend(handles, labels)
plt.title("protein")
```
